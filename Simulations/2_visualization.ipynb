## 1. Import Packages<a name="imports"></a>

```python
import numpy as np
import deeplenstronomy.deeplenstronomy as dl
```

```python
from deeplenstronomy.visualize import view_image, view_image_rgb
```

## 2. Visualize images<a name="see"></a>

### 2.2. `CONFIGURATION` 1 and 2<a name="see12"></a>

We are visualizing `CONFIGURATION` 1: Galaxy-Galaxy Lensing.

```python
high_cad_1_2_config1_images = np.load('high_cad_1_2_data/CONFIGURATION_1_images.npy')
```

```python
for i in np.arange(4):
    view_image(high_cad_1_2_config1_images[0][i])
```

```python
view_image_rgb(high_cad_1_2_config1_images[42], Q=10, stretch=10)
```

We are visualizing `CONFIGURATION` 2: Two Nearby Galaxies.

```python
high_cad_1_2_config2_images = np.load('high_cad_1_2_data/CONFIGURATION_2_images.npy')
```

### 2.2. `CONFIGURATION` 3<a name="see3"></a>

You will often have to tweak the `Q` and `stretch` values when using this function. Again, you can also pass `matplotlib.pyplot.imshow()` keyword arguments to the function, but keywords like `cmap` are discouraged to allow the `make_lupton_rgb()` to control the colors displayed.
